# Hands-on Python for ML

## Pandas DataFrame and Scikit-Learn

```
# Download excel file from Github, and convert it to a csv file
# Upload it onto the working directory

import pandas as pd

df = pd.read_csv('diabetes_data.csv')
df
```

```
# pd.head, tail, columns, index, len, shape
# index column 새로 만들어서 추가하기; 컬럼을 인덱스로 만들기
# ---- 불러올 때 부터 지정해 줄 수 있음 (index_col, header)
# 컬럼 추가할 때 맨 첫 줄로: 1) 만든 후에 옮기기, 
#                       2) 만들 때 부터 (df.insert(loc, column, value)
```

```
df['ID'] = range(1, len(df)+1)
df = df.set_index('ID') # This is not an 'inplace' method.
df
```

```
df = df.reset_index() 
# the previous 'ID' index is now in the first 'ID' column.
df
```

```
# subsetting: select columns and rows
# index, slicing - review (list, string, numpy)
# loc and iloc (indexer)
```

```
# Using iloc
df.iloc[:, 1:12]
```

```
# Using loc
df.loc[:, 'AGE':'Y']
```

```
# Get our final dataframe
df = df.loc[:, 'AGE':'Y']
df
```

```
# Split the dataset into Train and Test sets
nrow = len(df) # or df.shape[0]
test_ratio = 0.3
n_train = int(nrow * (1-test_ratio))

X_train = df.iloc[:n_train, :-1]
y_train = df.iloc[:n_train, -1]

X_test = df.iloc[n_train:, :-1]
y_test = df.iloc[n_train:, -1]

print('No. of train samples:', len(X_train))
print('No. of test samples:', len(X_test))
```

```
#0. Preprocessing- standardization
# Income 변수는 1원 단위로 하고 age는 1살 단위로 한다면?
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train, y_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor

#1. Create an empty regressors
lin_reg = LinearRegression()
ada_reg = AdaBoostRegressor()

#2. Fit the regressors using the train dataset
lin_reg.fit(X_train, y_train)
ada_reg.fit(X_train, y_train)
```

```
# Coefficients of lin_reg are in the 'coef_' attribute
print(lin_reg.coef_)
```

```
#3. Predict from X_test dataset

# Linear regression
y_train_pred_lin = lin_reg.predict(X_train)
y_test_pred_lin = lin_reg.predict(X_test)

# Adaptive boosting
y_train_pred_ada = ada_reg.predict(X_train)
y_test_pred_ada = ada_reg.predict(X_test)
```

```
#4. Evaluation
from sklearn.metrics import mean_absolute_error 

# Linear regression
print('Linear Regression:')
print('MAE in Training set: ', mean_absolute_error(y_train, y_train_pred_lin))
print('MAE in Test set: ', mean_absolute_error(y_test, y_test_pred_lin))

# Adaptive Boosting
print('Adaptive Boosting:')
print('MAE in Training set: ', mean_absolute_error(y_train, y_train_pred_ada))
print('MAE in Test set: ', mean_absolute_error(y_test, y_test_pred_ada))

#  We are highly likely to observe under-fitting with Linear regression
# and over-fitting with AdaBoost regression in this setting.
```
